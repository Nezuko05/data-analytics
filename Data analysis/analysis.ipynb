```python
df=sns.load_dataset("titanic")
df
```

```python
df.isnull().sum()
```

```python
avg_age = round(df["age"].mean())
print(avg_age)
```

```python
#filling missing values with average age

df["age"].fillna(avg_age,inplace=True)
df.head(10)
```

```python
#to remove

df.dropna(subset=["embarked"],inplace=True)
df.isnull().sum()
```
