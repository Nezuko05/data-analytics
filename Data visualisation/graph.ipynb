```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
sns.countplot(df, x="survived" , hue="sex",palette="coolwarm")
plt.title("count of survived vs Not survived by Gender")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
sns.scatterplot(df, x="age", y="fare", hue="survived",palette="viridis")
plt.title("Scatterplot of Age vs Fare by Survived")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.countplot(df,  y="class",palette="viridis")
plt.title("Count of Passenger by Class")
plt.xlabel("count")
plt.ylabel("Class")
plt.show()
```

```python
sns.barplot(df,x='age',y='fare',hue='survived',palette='viridis')
plt.title('Scatterplot of age vs fare by survival')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
sns.histplot(df,x='age',y='fare',hue='survived',palette='viridis')
plt.title('Scatterplot of age vs fare by survival')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
sns.pointplot(df,x='age',y='fare',hue='survived',palette='viridis')
plt.title('Scatterplot of age vs fare by survival')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
sns.boxplot(df,x='age',y='fare',hue='survived',palette='viridis')
plt.title('Scatterplot of age vs fare by survival')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
plt.figure(figsize=(8, 5))  
sns.barplot(df,x="sex", y="survived", palette="pastel")


plt.title("Survival Rate by Gender", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Survival Rate", fontsize=12)


plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
```

```python
survived_counts=df['survived'].value_counts()
px.pie(names=survived_counts.index, values=survived_counts.values,
       title='Survived Distribution', template='plotly_dark', width=500, height=500)
```

```python
px.scatter(df,x='age',y='fare', color='survived', title='Age vs Fare by Survival',
           template='plotly_dark', width=809)
```

```python

px.sunburst(df, path=['pclass','sex','survived'],
            title='Survived by class and gender', template='plotly_dark',
            width=800)
```

```python
px.histogram(df, x='age',color='sex', facet_col='survived',
              title='age distribution by gender and survived',
              
              template='plotly_dark',width=800)
 
```
